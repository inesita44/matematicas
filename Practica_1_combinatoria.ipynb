 - Intervienen todas las bolas, hay 7 bolas y se sacan las 7 bolas.
 - Se repiten elementos, es decir pueden sacarse bolas más de una bola del mismo color
* Los casos que se obtienen con las permutaciones con repetición son $PR_{n}^{a,b,c...}=\frac{n!}{a!b!c!...}$ que para este ejemplo son:

```python
from math import factorial
from sympy.utilities.iterables import multiset_permutations
BOLSA='RRRRVVV'
m = len(BOLSA)
lista = [''.join(i) for i in multiset_permutations(BOLSA)]
print(f"El número de maneras que se pueden extraer una a una las bolas de la urna es: {len(lista)}.")
```

>#### **Historia**

En las matemáticas, el triángulo de Pascal es una representación de los
coeficientes binomiales ordenados en forma de triángulo. Es llamado así en
honor al filósofo y matemático francés **Blaise Pascal**, quien introdujo esta
notación en **1654**, en su **Traité du triangle arithmétique**. Si bien las
propiedades y aplicaciones del triángulo fueron conocidas con anterioridad al
tratado de Pascal por matemáticos **indios, chinos, persas, alemanes e italianos**,
fue Pascal quien desarrolló muchas de sus aplicaciones y el primero en organizar
la información de manera conjunta.

>#### **Su forma**

El triángulo de Pascal es un **triángulo de números enteros, infinito y simétrico** que empieza con un 1 en la primera fila, y en las filas siguientes se van colocando números de forma que cada uno de ellos sea la suma de los dos números que tiene encima. Las **diagonales** que empiezan desde el «1» situado en la cabeza del triángulo valen siempre 1

>#### **Combinatoria y el triangulo de Pascal**

La construcción del triángulo está relacionada con los coeficientes binomiales según la fórmula (también llamada regla de
Pascal) combinatoria. Esta fórmula o regla explica que los coeficientes (nodos del "árbol") de una fila dada del triángulo, se
pueden calcular con la fórmula combinatoria de combinaciones de **n** elementos de **k en k** ; dónde **n**  es la fila - 1 y **k** la posición en la fila.
